```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = 'https://www.ola.org/en/legislative-business/house-documents/parliament-42/session-1/2020-07-21/hansard'
```

```python
page = urlopen(url)
```

```python
type(page)
```

```python
html_bytes = page.read()
```

```python
html = html_bytes.decode("utf-8")
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```
